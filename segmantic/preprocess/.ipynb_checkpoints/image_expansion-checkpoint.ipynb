```python
def makeFile(path):
    if not os.path.exists(path):
        os.mkdir(path)
```

```python
def convertAnnotation(pixel):
    pixel[pixel>0] = 255
    return pixel
```

```python
def setName(ind):
    if ind < 10:
        name = '0000' + str(ind)
    elif ind < 100:
        name = '000' + str(ind)
    elif ind < 1000:
        name = '00'+ str(ind)
    elif ind < 10000:
        name = '0' + str(ind)
    else:
        name = str(ind)
    return name
```

### 针对human parsing处理

```python
street_img_name = os.listdir(street_img_path)
len(street_img_name)
```

```python
def getImage(allname, show=False):
    img = ann = 0
    name = os.path.splitext(allname)[0]
    img_path = street_img_path + name + '.jpg'
    ann_path = street_ann_path + name + '.png'
    img = cv2.imread(img_path)
    h, w, _ = img.shape
    if h >= 512 and w >= 400:
        if show:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        ann = cv2.imread(ann_path)
        ann = convertAnnotation(ann)
        ann = cv2.medianBlur(ann, 5) 
        ann = cv2.medianBlur(ann, 3)
    return img, ann
```

```python
img, ann= getImage(street_img_name[4], show=True)
if not isinstance(ann, int):
    fig = plt.figure(figsize=(12, 8))
    ax1 = fig.add_subplot(1, 2, 1)
    label = np.logical_not(np.logical_and(img, ann))
    img[label] = 255
    ll = plt.imshow(img)
    ax2 = fig.add_subplot(1, 2, 2)
    zz = plt.imshow(ann)
    plt.show()
```

### 针对衣服服饰处理

```python
import scipy.io as scio
```

```python
cloth_name = os.listdir(cloth_img_path)
len(cloth_name)
```

```python
def getHeadImage(allname, show=False):
    img = ann = 0
    name = os.path.splitext(allname)[0]
    if int(name) <= 1004:
        img_path = cloth_img_path + name + '.jpg'
        ann_path = cloth_pann_path + name + '.mat'
        img = cv2.imread(img_path)
        if show:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        item = scio.loadmat(ann_path)['groundtruth']
        high, width = item.shape
        ann = np.zeros((high, width, 3), dtype=np.uint8)
        for i in range(3):
            ann[:, :, i] = item
        ann = convertAnnotation(ann)
    return img, ann
```

```python
img3, ann3= getHeadImage(cloth_name[44], show=True)
if not isinstance(img3, int):
    fig3 = plt.figure(figsize=(12, 8))
    ax1 = fig3.add_subplot(1, 2, 1)
    ll = plt.imshow(img3)
    ax2 = fig3.add_subplot(1, 2, 2)
    zz = plt.imshow(ann3)
    plt.show()
```

```python
cloth_save = 'datasets/cloth/'
cloth_save_img = cloth_save + 'image/'
cloth_save_ann = cloth_save + 'annotation/'
makeFile(cloth_save)
makeFile(cloth_save_img)
makeFile(cloth_save_ann)
```

```python
i = 0
for cloth in cloth_name:
    img3, ann3= getHeadImage(cloth)
    if not isinstance(img3, int):
        cv2.imwrite(cloth_save_img + setName(i) + '.jpg', img3)
        cv2.imwrite(cloth_save_ann + setName(i) + '.png', ann3)
        i += 1
```

### 过滤图片并且重新命名 

```python

len(street_name)
```

```python
import shutil

def moveAndRename(path):
    street_name = os.listdir(path + 'annotations/')
    for i, name in enumerate(street_name):
        base_name = os.path.splitext(name)[0]
        img = path + 'images/' + base_name + '.jpg'
        des_img = path + 'image/' + setName(i) + '.jpg'
        ann = path + 'annotations/' + name
        des_ann = path + 'annotation/' + setName(i) + '.png'
        
        shutil.move(img, des_img)
        shutil.move(ann, des_ann)
```

```python
moveAndRename('datasets/street/')
```

```python
def rename(path):
    i = 1004
    street_name = os.listdir(path + 'annotation/')
    for name in street_name:
        i += 1
        base_name = os.path.splitext(name)[0]
        img = path + 'image/' + base_name + '.jpg'
        des_img = 'datasets/street/image/' + setName(i) + '.jpg'
        ann = path + 'annotation/' + name
        des_ann = 'datasets/street/annotation/' + setName(i) + '.png'
        
        shutil.move(img, des_img)
        shutil.move(ann, des_ann)
```

```python
rename('datasets/persona/')
```
